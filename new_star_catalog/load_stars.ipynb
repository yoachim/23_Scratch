```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
import healpy as hp
from rubin_sim.utils import healbin
```

```python
stars = pd.read_csv('stars_stars_1.csv')
```

```python
stars
```

```python
ack = healbin(stars['ra'], stars['dec'], stars['ra']*0+1, 64, reduce_func=np.sum)
```

```python
hp.mollview(ack, max=100)
```

```python
good = np.where(ack >= 0)[0]
ack[good].min()
```

```python
_temp = plt.hist(ack[good], bins=np.arange(30)-0.5)
```
