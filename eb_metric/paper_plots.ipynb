```python


import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import rubin_sim.maf as maf
import healpy as hp
import os
```

```python
# from http://astro-lsst-01.astro.washington.edu:8080/fbs_db/fbs_2.1/baseline/baseline_v2.1_10yrs.db
filename = '/Users/yoachim/Downloads/baseline_v2.1_10yrs.db'
run_name = os.path.basename(filename.replace('.db', ''))
```

```python
bundle_list = []

slicer = maf.HealpixSlicer()
metric = maf.CountMetric(col='night')
sql = ''

plot_dict={'title': run_name, 'xlabel': 'Number of Observations', 'color_min': 100, 'color_max': 1000,
          'x_max': 1000}

bundle_list.append(maf.MetricBundle(metric, slicer, sql, plot_dict=plot_dict, run_name=run_name))
```

```python
bundle_dict = maf.make_bundles_dict_from_list(bundle_list)
bg = maf.MetricBundleGroup(bundle_dict, filename, out_dir='temp',results_db=None)
```

```python
bg.run_all()
bg.plot_all(closefigs=False)
```
