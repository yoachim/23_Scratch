```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import healpy as hp
from rubin_sim.scheduler.features import  NObservations
from rubin_sim.scheduler.utils import HpInLsstFov, read_fields
from rubin_sim.scheduler.thomson import fib_sphere_grid
from rubin_sim.utils import ra_dec_from_xyz
```

```python
# data file from http://neilsloane.com/icosahedral.codes/index.html
#points = np.genfromtxt('icover.3.5292.23.0.txt')
points = np.genfromtxt('neilsloane/neilsloane.com/ICOSC/icover.3.5292.23.0.txt')
x = points[0::3]
y = points[1::3]
z = points[2::3]
```

```python
ra,dec = ra_dec_from_xyz(x,y,z)
```

```python
nside = 2048
pointing2hpindx = HpInLsstFov(nside=nside)
```

```python

feature_new = NObservations(nside=nside)

for ra_val,dec_val in zip(np.radians(ra),np.radians(dec)):
    indx = pointing2hpindx(ra_val, dec_val)
    feature_new.add_observation(None, indx=indx)
    
```

```python
hp.mollview(feature_new.feature, rot=(0,0,-30))
```
