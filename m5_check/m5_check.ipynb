```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
def classic_m5(m, m5):
    snr = 5.0 * 10.0 ** (-0.4 * (m - m5))
    return snr
```

```python
def new_m5(m,m5, gamma=0.039):
    # gamma varies per band, but is fairly close to this value
#    rgamma = 0.039
    xval = np.power(10, 0.4 * (m - m5))
    snr = 1 / np.sqrt((0.04 - gamma) * xval + gamma * xval * xval)
    return snr
```

```python
m=np.arange(15,30,0.2)
m5 = 24.

cm5 = classic_m5(m,m5)
nm5 = new_m5(m,m5)
```

```python
plt.semilogy(m, cm5, label='classic')
plt.plot(m, nm5, label='new')
plt.legend()
plt.xlabel('m')
plt.ylabel('SNR')
```
